# Извлечение картинок из файла PDF

```python
import fitz
import os

if not os.path.isdir('images'):
    os.mkdir('images')
    
file = 'file.pdf'

pdf = fitz.open(file)

for i in range(len(pdf)):
    image_list = pdf.get_page_images(i)

    for image in image_list:
        xref = image[0]
        pix = fitz.Pixmap(pdf, xref)
        if pix.n < 5:
            pix.save(os.path.join('images', f'{xref}.jpeg'))
        else:
            pix1 = fitz.open(fitz.csRGB, pix)
            pix1.save(os.path.join('images', f'{xref}.jpeg'))
            pix1 = None
        pix = None
```

# Удаление заднего фона из картинок

```python
from rembg.bg import remove
import numpy as np
import io
from PIL import ImageFile, Image
```

```python
place = 'images'
if not os.path.isdir('images_without_bg'):
    os.mkdir('images_without_bg')
for pict in os.listdir(place):
    if pict.endswith('.png') or pict.endswith('.jpg') or pict.endswith('.jpeg'):
        print(f'[+] Удаляю фон: "{pict}"...')
        output = remove(Image.open(os.path.join(place, pict)))
        output.save(os.path.join('images_without_bg', f'{pict.split(".")[0]}.png'))
    else:
        continue
```

# Извлечение таблиц из файла PDF

```python
import tabula
import pandas as pd
from tabula.io import read_pdf

if not os.path.isdir('tables'):
    os.mkdir('tables')

pdf_in = 'file.pdf'

PDF = tabula.read_pdf(pdf_in, pages = 'all' , encoding = 'Windows-1251', stream = True)

for i in range (len(PDF)):
    PDF[i].to_excel(os.path.join('tables', f'{i}.xlsx'), index=False) 

print("Done")
```
